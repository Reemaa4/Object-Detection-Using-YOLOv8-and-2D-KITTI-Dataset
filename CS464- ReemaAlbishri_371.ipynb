```python
# (Testing process using the model made by the YOLO training "best.pt")
from ultralytics import YOLO
import os
import random
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# YOLO trained model path 
model = YOLO(r"C:\Users\Lapto\Downloads\Deep Midterm\runs\training_results5\weights\best.pt")

# Test images dataset
test_image_dir = r"C:\Users\Lapto\Downloads\Deep Midterm\data_object_image\testing\image_2"

# save all the images 
image_files = [f for f in os.listdir(test_image_dir) if f.endswith('.png')]


# 4. Select 20% of the images
num_images_to_select = int(len(image_files) * 0.2)  
selected_images = random.sample(image_files, num_images_to_select) # Randomly select the images
```

```python
all_preds = []
all_gts = []  

# Stat testing using the model "best.pt" and save the result 
for img_file in selected_images:
    img_path = os.path.join(test_image_dir, img_file)
    
    results = model(img_path, save=True)
    
    pred_boxes = results[0].boxes 

    preds_classes = pred_boxes.cls.cpu().numpy().astype(int)  
    preds_conf = pred_boxes.conf.cpu().numpy()  
    all_preds.extend(zip(preds_classes, preds_conf)) 

    
    gt_boxes = [] 
    all_gts.append(gt_boxes)
```
